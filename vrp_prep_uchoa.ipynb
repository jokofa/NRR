### Code for loading and converting the Uchoa benchmark data

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
```

```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from lib.problem.generator import RPGenerator

sns.set()
```

```python
PART = 3
ROOT_DIR = f"./data/CVRP/benchmark/uchoa/"
fstr = f"raw/n{PART}"
DATA_DIR = os.path.join(ROOT_DIR, fstr)
DATA_DIR
```

```python
files = os.listdir(DATA_DIR)
files.sort()
files
```

```python
instances = []
for f in files:
    load_pth = os.path.join(DATA_DIR, f)
    inst = RPGenerator.load_dataset(load_pth)
    print(load_pth)
    print(inst)
    instances.append(inst[0])
```

```python
instances
```

```python
save_pth = os.path.join(ROOT_DIR, f"n{PART}.dat")
RPGenerator.save_dataset(instances, filepath=save_pth)
```

```python
spth = os.path.join(ROOT_DIR, f"n{PART}.dat")
instances = RPGenerator.load_dataset(spth, allow_pickle=True)
instances
```

```python
COLS = ["x_coord", "y_coord", "demand"]
i = 0
for inst in instances:
    df_ = pd.DataFrame(
        data=np.concatenate((inst.coords, inst.demands[:, None]), axis=-1),
        columns=COLS
    )
    sns.relplot(x="x_coord", y="y_coord", size="demand",
                sizes=(10, 100), alpha=.5, palette="muted",
                height=6, data=df_)
    plt.scatter(inst.coords[0, 0], inst.coords[0, 1], c="red", marker="*", s=100)
    i += 1
    if i > 10:
        break
```
